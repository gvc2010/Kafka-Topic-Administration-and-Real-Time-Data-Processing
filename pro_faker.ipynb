```python
from kafka import KafkaProducer
from faker import Faker
import json
import time

fake = Faker()

#funcao para criar dados(fake)

def get_customer():
    return {
        'nome' : fake.name(),
        'profissao' : fake.job(),
        'endereco' : fake.address(),
        'nascimento' : fake.year()
    
    }

SERVER = 'guilherme.venturini-broker'
PORT = '29092'
TOPIC = 's7projeto'
SERVER_INFO = SERVER + ':' + PORT

producer = KafkaProducer(bootstrap_servers=[SERVER_INFO])


while True:
    msg = get_customer()
    data = f"FAKE,{json.dumps(msg)}"
    print(f"Sendig Fake Data: {data}")
    producer.send(TOPIC, value=data.encode('utf-8'), partition=1)
    time.sleep(2)
    
```
