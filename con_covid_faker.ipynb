```python
from kafka import KafkaConsumer, TopicPartition
import json

SERVER = 'guilherme.venturini-broker'
PORT = '29092'
TOPIC = 's7projeto'
SERVER_INFO = SERVER + ':' + PORT

consumer = KafkaConsumer(bootstrap_servers=[SERVER_INFO],
                         auto_offset_reset='earliest',
                         enable_auto_commit=True,
                         group_id='my-group',
                         value_deserializer=lambda x: x.decode('utf-8'))

#assignando as particoes especificas ao consumidor

partitions = [TopicPartition(TOPIC, 0), TopicPartition(TOPIC, 1)]
consumer.assign(partitions)


for message in consumer:
    value = message.value
    if value.startswith("COVID,"):
        print(f"COVID data: {value[6:]}")
    elif value.startswith("FAKE,"):
        fake_data = json.loads(value[5:])
        print(f"Fake Data: {fake_data}")
    else:
        print(f"Formato Desconhecido: {value}")
              
        
    
    
```
